Limpieza de datos

```python
def info_general_data_books(data_books):
  print('Información general del DataFrame')
  print(f'Cantidad de filas y columnas: {data_books.shape}')
  print("-"*50)
  print(f'Cantidad de datos nulos por columna: \n{data_books.isnull().sum()}')
  print("-"*50)
  print(f'Cantidad de datos únicos por columna: \n{data_books.nunique()}')
  print("-"*50)
  print(f'Tipos de datos por columna: \n{data_books.dtypes}')

info_general_data_books(data_books)

#revisar tipos y cantidades de datos
```

```python
data_books_current = data_books.copy()
data_books_current.head()
#s crea una copia del dataset por seguridad
```

```python
data_books_current['publishedDate'] = pd.to_datetime(data_books_current['publishedDate'], errors='coerce')
#pasar fecha de publicación a tipo de dato fecha
```

```python
print("Nuevos tipos de datos")
display(data_books_current.dtypes)
#revisar que el cambi de tipo de dato este correcto
```

```python
data_fillna = {
    'Title': "Anonimo",
    'authors': "Anonimo",
    'description': "Aún estamos trabajando en la descripción de este título",
    'publisher': "Editorial independiente",
    'categories': "Categoría no fija",
    'review/helpfulness': "Sin calificar reviews",
    'review/summary': "Se el primero en calificar este título",
    'previewLink': "Error 404 - No encontrado",
    'User_id': "Anonimo",
    # Asignamos una cadena para los nombres faltantes

# Generar fechas aleatorias para las columnas de fechas
    'publishedDate': pd.to_datetime(np.random.uniform(data_books_current['publishedDate'].min().value, data_books_current['publishedDate'].max().value)).normalize(),


    # Usar la media para completar valores numéricos faltantes
    'Price': data_books_current['Price'].mean(),  # Sacamos media para valores faltantes
    'review/score': data_books_current['review/score'].mean(),  # Sacamos media para completar el puntaje
    'ratingsCount': data_books_current['ratingsCount'].mean(),  # Sacamos media para sacar el
}
```

```python
data_books_current.fillna(data_fillna, inplace=True)
# Verificar la cantidad de nulos restantes
print(data_books_current.isnull().sum())
```

```python
data_books_current = data_books_current.drop(columns=['infoLink','profileName','review/text','image'])
#Se eliminan las columnas que no se van a usar
print(data_books_current.columns)
```
